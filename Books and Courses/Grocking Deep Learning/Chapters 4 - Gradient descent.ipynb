# Hot and Cold learning method, bacics

Посчитать ошибку и при шаге в + и при шаге в -. Сравнить эти ошибки, сделать шаг в ту сторону где ошибка меньше.

Задача обучения нейронных сетей, по сути, является задачей поиска (оптимальных параметров).

```python
def neural_network_1(input, weight):
    prediction = input * weight
    return prediction

weight = 0.1
lr = 0.01
input = 8.5
correct_answer = 1

# исходная ошибка
pred = neural_network_1(input, weight)
square_error = (pred - correct_answer) ** 2
print('Starting error = ', square_error)

# ошибка при шаге в положительную сторону
pred_up = neural_network_1(input, weight+lr)
square_error_up = (pred_up - correct_answer) ** 2
print('+lr error = ', square_error_up)

# ошибка при шаге в отрицательную сторону
pred_down = neural_network_1(input, weight-lr)
square_error_down = (pred_down - correct_answer) ** 2
print('-lr error = ', square_error_down)

# если одна из новых ошибок меньше исходной то...
if square_error > square_error_up or square_error > square_error_down:
    if square_error_down < square_error_up:
        weight -= lr  # сделать шаг в отрицательную сторону
        print('Lr Down')
    if square_error_up < square_error_down:
        weight += lr  # сделать шаг в положительную сторону
        print('Lr Up')
```

# Hot and Cold learning method, actual learning algorithm

Не очень эффективный метод, т.к. на каждом шаге мы считаем три предсказания.

Плюс, из-за того что goal почти никогда не равен N * Lr, мы никогда не снизим ошибку до нуля. А если Lr окажется слишком большим мы можем и вовсе никогда и не начать снижать ошибку, "прыгая" вокруг (дефолтная проблема статического Lr). Если Lr очень маленький - время обучения может устремиться в бесконечность.

```python
weight = 0.5
input = 0.5
goal = 0.8  # correct answer
lr = 0.001
print('Goal = 0.8')

for i in range(1500):
    prediction = input * weight
    sqr_error = (prediction - goal) ** 2
    print("itteration: ", i, "\n Square error = ", sqr_error, '\n Prediction = ', prediction)
    
    up_pred = input * (weight + lr)
    up_err = (up_pred - goal) ** 2
        
    down_pred = input * (weight - lr)
    down_err = (down_pred - goal) ** 2
    
    if down_err < up_err:
        weight -= lr
        print('Weight Down')
        
    if up_err < down_err:
        weight = weight + lr
        print(' Weight Up')
```

# Gradient descent

1) Считаем на сколько ошибся каждый нейрон (абсолютная ошибка, НЕ квадратная! Здесь нам нужен знак).

2) Смотрим на сколько каждый из весов заставил нейрон ошибиться (умножаем вес на ошибку нейрона - это количественная оценка влияния веса на ошибку нейрона, взвешенная дельта).

3) Корректируем каждый вес на его взешенную дельту умноженную на шаг градиентного спуска Альфа.

Таким образом мы понемного изменяем каждый вес таким образом, чтобы ответ сети был ближе и ближе к целевому значению.

```python
# упрощённый пример (очень сильно упрощённый)

weight = 0.5
goal = 0.8
input = 0.5

for i in range(120):
    pred = input * weight
    sqr_error = (pred - goal) ** 2
    gradient = (pred - goal) * input
    weight -= gradient  # -= потому что градиент показывает направление роста, а нам наоборот нужно уменьшать
    print('Itteration: ', i, '\n Prediction = ', pred, '\n Sqr_error = ', sqr_error, '\n Gradient = ', gradient)
```

```python
# более реалистичный пример градиентного спуска

goal = 1
input = 8.5
weight = 0.1
alpha = 0.01  # gradient descent's step size 

for i in range(50):
    pred = input * weight
    error = (pred - goal) ** 2  # error shows how much we missed the target
    delta = pred - goal  # delta shows how much the particular neuron has missed (right here we have only one neuron)
    derivative = delta * input  # shows how much the weight caused the neuron to miss, это производная от функции ошибки по весу
    weight -= derivative * alpha  # adjusting the weight so it will cause less miss for the neuron
    print('Itteration: ', i, '\n Goal = 1\n Prediction = ', pred, '\n Error = ', error,
         '\n Weight = ', weight)
```

## Почему и как это работает?

Возьмём функцию ошибки $error = (pred - goal)^ 2$ и раскроем функцию pred. pred возвращает значение равное $input * weight$. Значит функцию error можно записать как $error = ((input * weight) - goal)^2$.

Однако input и goal - это КОНСТАНТЫ. Они задаются в начале обучения и НЕ изменяются. Это означает что error является функцией от weight: $error(weight) = ((input * weight) - goal)^2$. Больше НИКАКИХ переменных в этой функции нет.

Это значит что есть некая зависимость между error и weight, и во множестве значений принимаемых функцией $error(weight)$ есть точка в которой error минимальна, ей соответствует некое значение weight которое мы и пытаемся найти.

#### Производная - это чувствительность между двумя переменными. Она показывает как сильно и в какую сторону изменяется одна переменная при изменении второй.

Мы находим производные между всеми весами и ошибкой. Затем сдвигаем значения всех весов в противоположную сторону. Именно так обучается нейронная сеть.
