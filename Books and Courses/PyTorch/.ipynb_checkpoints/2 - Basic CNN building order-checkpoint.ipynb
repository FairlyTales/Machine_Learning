# Create the neural network class

## Create new class which extends the PyTorch class "torch.nn.Module".

We use "super().__init__()" to initialize the "__init__()" of the "nn.Module" in our class (we are required to use super, without it nn.Module`s "__init__()" won`t be activated and we won`t get the needed methods and arguments).

## Specify the layers of our network.

Main rule is: the quantity of OUTs in a previous layer is a quantity of INs for the next layer.

"in_channels" in "self.conv1" must be a quantity of channels in the training set (we have grayscale, so we set it to 1).

"out_features" in "self.out" must be a quantity of output classes in the set (we have 10 classes of images in this set, so we set it to 10).

## Specify the forward method

Метод определяющий как наши данные будут перемещаться по сети (сейчас речь только о forward propagation).

В нём мы указываем каким образом и в каком порядке какие операции будут проводиться над данными.

return этого метода возвращает тензор полученный после одного прохода данных по сети.

```python
class Net(nn.Module):
    def __init__(self):        
        super().__init__()
        
        # create the layers
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=12, kernel_size=5)        
        self.fc1 = nn.Linear(in_features=12 * 4 * 4, out_features=120) # 12*4*4 = 12 "инпутов" 4 на 4 пикселя
        self.fc2 = nn.Linear(in_features=120, out_features=60)
        self.out = nn.Linear(in_features=60, out_features=10)
        
    def forward(self, t):
        t = self.conv1(t)
        t = F.relu(t) # set an activation function
        t = F.max_pool2d(t, kernel_size=2, stride=2) # pooling layer
        
        t = self.conv2(t)
        t = F.relu(t)
        t = F.max_pool2d(t, kernel_size=2, stride=2)
        
        t = t.reshape(-1, 12 * 4 * 4) # flattening the tensor before forwarding it to fc1
        t = self.fc1(t)
        t = F.relu(t)
        
        t = self.fc2(t)
        t = F.relu(t)
        
        t = self.out(t)
        
        return t
```

# Train the neural network

## Create a neural network class object, optimizer

## Set the number of epochs

## Create an epochs cycle 

## Create a batches cycle inside the epochs cycle

## Specify loss function

```python
# set epochs and learning rate (batch_size settings is in the cell above)
Epoch = 3
learning_rate = 0.01

# create a network from our network class & send it to GPU
network = Net().to(device)

# create an optimizer and pass to it parameters of the network, set the learning rate
optimizer = optim.Adam(network.parameters(), lr=learning_rate)

# count the number of correct predictions based on predictions & labels tensors
def get_num_correct(predictions, labels):
    return predictions.argmax(dim=1).eq(labels).sum().item()

# send training data and network architecture to TensorBoard
comment = f' batch_size={batch_size} learning_rate={learning_rate}' # to identify the training runs
tb = SummaryWriter(comment=comment) # create a SummaryWriter object (used to prepare data for tensorboard)
tb.add_image('images', grid) # send our grid (check the Ch.2 Prepare the data, visualization) of images to tensorboard
tb.add_graph(network, images.to(device)) # send our network architecture and input data to TensorBoard

# main training cycle
for epoch in tqdm(range(Epoch), desc='Epochs'):
    # counters for loss and correct predictions
    total_loss = 0
    total_correct = 0

    for batch in tqdm(train_loader,  desc='Batch'):  # get the batch
        images, labels = batch  # assign image-label pair to their own object
        images = images.to(device) # transfer inputs to the GPU
        labels = labels.to(device) # transfer inputs to the GPU
        predictions = network(images)  # pass batch

        # set the loss function
        loss = F.cross_entropy(predictions, labels)  # calculate loss

        # backpropagation
        optimizer.zero_grad()  # reset the gradients to prevent them from stacking with previous
        loss.backward()  # calculate gradients
        optimizer.step()  # update weights

        # update counters
        total_loss += loss.item() * batch_size # * batch_size - to scale loss calc. for comparing them
        total_correct += get_num_correct(predictions, labels)
        
    
    # send scallar data to TensorBoard
    tb.add_scalar('Loss', total_loss, epoch)
    tb.add_scalar('Correct', total_correct, epoch)
    tb.add_scalar('Accuracy', total_correct / len(train_set), epoch)
    
    # send histograms to TensorBoard
    for name, weight in network.named_parameters():
        tb.add_histogram(name, weight, epoch)
        tb.add_histogram(f'{name}.grad', weight.grad, epoch)
        
    # print results of the each epoch
    print('Epochs: ', epoch, ', Correct predictions: ',
          total_correct, ', Loss: ', total_loss)
    

# close the connection to the TensorBoard, saving the results
tb.close() 


# save the network parameters
PATH = './cifar_net.pth'
# torch.save(network.state_dict(), PATH)

# or save the whole network
# torch.save(net, PATH)

# to load a network state:
# net.load_state_dict(torch.load(PATH))

# to load a full network:
# net = torch.load(PATH)
```

```python
@torch.no_grad() # turn gradient tracking off for the next operations to reduce computation cost
def get_all_preds(model, loader):
    """Returns all predictions from the network in form of a tensor"""
    
    all_preds = torch.tensor([])
    all_preds = all_preds.cuda()
    for batch in loader:
        images, labels = batch
        images = images.to(device) # transfer inputs to the GPU
        labels = labels.to(device)
        preds = model(images)
        all_preds = torch.cat((all_preds, preds), dim=0) # torch.cat - concatenation
        
    return all_preds

# create loader for predictions
with torch.no_grad(): # turn gradient tracking off for the next operations to reduce computation cost
    prediction_loader = torch.utils.data.DataLoader(train_set, batch_size=1000)
    train_preds = get_all_preds(network, prediction_loader) # our prediction tensor

# create a list of actual dataset labels, which correspods to the numbers from 0 to 9 in the dataset
# check the dataset's readme in order to do it correctly
dataset_labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
                  'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# calculate and plot the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(train_set.targets.cpu(), train_preds.argmax(dim=1).cpu()) # argmax(dim=1) чтобы из 10 вероятностей выбрать одну, наибольшую
plt.figure(figsize=(15, 10))
sns.heatmap(cm, square=True, annot=True, annot_kws={"size": 16}, cmap='copper', fmt='d',
            xticklabels=dataset_labels, yticklabels=dataset_labels)
plt.xlabel("Predicted")
plt.ylabel("True");
```
