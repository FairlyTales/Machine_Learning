# Basic Operations on Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('dark_background')

# Accessing and Modifying pixel values
img = cv2.imread('D:/Stuff on HDD/Pictures/OLD wallapers/__pyrrha_nikos_rwby_drawn_by_dishwasher1910__294757036e96b0ae5f12a971f4d42564.jpg', 1)
# flag=1 - colored image, any transparency of image will be neglected
# flag=0 - grayscale image, any transparency of image will be neglected
# flag=-1 - unchanged image, with alpha channel (transparency)

# show image
"""
cv2.imshow('image', img)
cv2.waitKey(6000)
cv2.destroyAllWindows()
"""

# values of the pixel in Blue Green Red
px = img[100, 100]
print (px)

# accessing only blue pixel (blue=0, green=1, red=2)
blue_px = img[100, 100, 0]
blue_px

# modify the pixel values
img[100,100] = [255,255,255]
print (img[100,100])

# another way to acces the values
# accessing RED value
img.item(10,10,2)

# modifying RED value
img.itemset((10,10,2),100)
img.item(10,10,2)
```

```python
# Accessing Image Properties
print(img.shape) # 3 - number of "color matrices"
print(img.size) # total number of pixels
print(img.dtype) # very important thing for debugging, lot`s of OpenCV-Python error comes from invalid datatype

# Image ROI (Region of Interest)
piece = img[100:300, 20:400] # cut a piece of the image, basic NumPy indexing

"""
cv2.imshow('image', piece)
cv2.waitKey(6000)
cv2.destroyAllWindows()
"""

# Splitting Image Channels
b, g, r = cv2.split(img)

# cv2.split() is a "heavy" function, so better use NumPy indexing, which is faster:
b = img[:, :, 0]
g = img[:, :, 1]
r = img[:, :, 2]

# Merging Image Channels
img = cv2.merge((b,g,r))

# Borders for Images (Padding) with cv2.copyMakeBorder() method
# for the arguments read documentation. Here is a sample:
BLUE = [255,0,0]
img1 = img
replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')

# plt.rcParams["figure.figsize"] = (200,13)
plt.show();
```

# Bitwise operations - and, or, not, xor

```python
img1 = cv2.imread("drawing_1.png")
cv2.imshow("img1", img1)
cv2.waitKey(2000)
cv2.destroyAllWindows()

img2 = cv2.imread("drawing_2.png")
cv2.imshow("img2", img2)
cv2.waitKey(2000)
cv2.destroyAllWindows()

bit_and = cv2.bitwise_and(img2, img1) # bitwise conjunction of the two arrays
cv2.imshow("bit_and", bit_and)
cv2.waitKey(2000)
cv2.destroyAllWindows()

bit_or = cv2.bitwise_or(img2, img1) # bitwise disjunction of two arrays or an array
cv2.imshow("bit_or", bit_or)
cv2.waitKey(2000)
cv2.destroyAllWindows()

bit_xor = cv2.bitwise_xor(img1, img2) # bit-wise "exclusive or" operation on two arrays
cv2.imshow("bit_xor", bit_xor)
cv2.waitKey(2000)
cv2.destroyAllWindows()

bit_not = cv2.bitwise_not(img1) # Inverts every bit of an array
cv2.imshow("bit_not", bit_not)
cv2.waitKey(2000)
cv2.destroyAllWindows()

bit_not2 = cv2.bitwise_not(img2) # Inverts every bit of an array
cv2.imshow("bit_not2", bit_not2)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

# Code Speed Measurement and Improvement Techniques
OpenCV functions are optimized using SSE2, AVX etc. If our system (PC) supports these features, we should exploit them (almost all modern day processors support them). It is enabled by default while compiling. So OpenCV runs the optimized code if it is enabled, else it runs the unoptimized code. 

```python
print(cv2.useOptimized()) # default optimization setting is ON
cv2.setUseOptimized(False) # turn optimization OFF
print(cv2.useOptimized())
cv2.setUseOptimized(True) # turn optimization ON
print(cv2.useOptimized())
```

```python
# Measuring Performance in IPython
img111 = cv2.imread('opencv_logo.png', 0) # convert to grayscale, because cv2.countNonZero(img1)
%timeit cv2.countNonZero(img111)          # work only with src – single-channel array
%timeit z = np.count_nonzero(img111)

# OpenCV is ~40 times faster, use only them (numpy is much slower)
```

# Performance Optimization Tips

THE RIGHT WAY TO WRITE: at first try to implement the algorithm in a simple manner. Once it is working, profile it, find the bottlenecks and optimize them.

Also consider this:

        Avoid using loops in Python as far as possible, especially double/triple loops etc. They are inherently slow.
        Vectorize the algorithm/code to the maximum possible extent because Numpy and OpenCV are optimized for vector operations.
        Exploit the cache coherence.
        Never make copies of array unless it is needed. Try to use views instead. Array copying is a costly operation.

Even after doing all these operations, if your code is still slow, or use of large loops are inevitable, use additional libraries like Cython to make it faster.
