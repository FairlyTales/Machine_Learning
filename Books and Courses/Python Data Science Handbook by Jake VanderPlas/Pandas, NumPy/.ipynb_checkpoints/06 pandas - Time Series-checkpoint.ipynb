```python
# начнем с родных питоновских методов работы со временем

from datetime import datetime
datetime(year=2019, month=7, day=4)

from dateutil import parser
date = parser.parse('4th of July, 2019')
date.strftime('%A') # читай про эту команду в документации к datetime - strftime
# в питоне есть много удобных функций для "времени", но нет векторизации, т.е.
# с его временными форматами неудобно работать
```

```python
import numpy as np
date = np.array('2019-07-04', dtype=np.datetime64) # именно такой формат ввода
date + 4 # поддерживает векторные операции

# datetime64 хранит ограниченное количество данных, при увеличении разрешения
# (сотые, тысячные доли мили/наносекунд и т.д.) максимальный объём для кодирования
# информации понижается
np.datetime64('2015-07-04') # day-based datetime
np.datetime64('2015-07-04 12:00') # minute-based datetime
np.datetime64('2015-07-04 12:59:59.50', 'ns') # nanosecond-based time
# оптимальным разрешением является ns - норм разрешение и около 292 лет объема для записи

# в numpy нет многих удобных функций datetime и dateutil
```

```python
# в питоне есть удобные функции для работы со временем (в dateutil и datetime),
# но нет векторных операций. В numpy есть векторизация, но нет многих удобных
# функций. В pandas есть все, поэтому его и будем использовать

import pandas as pd

date = pd.to_datetime('4th of July, 2019')
date # pandas использует объект Timestamp для работы со временем
```

```python
# задаем индексы по Timestamp - удобная штука
index = pd.DatetimeIndex(['2014-07-04', '2014-08-04',
                          '2015-07-04', '2015-08-04'])
data = pd.Series([0, 1, 2, 3], index=index)
data
data['2014-07-04':'2015-07-04']
data['2015'] # по таким индексам удобно выделять данные
```

```python
# pd.to_datetime используется для создания pandasовских временных объектов
# основанных на np.datetime64. Этот метод принимает много вариантов ввода:
dates = pd.to_datetime([datetime(2015, 7, 3), '4th of July, 2015',
                       '2015-Jul-6', '07-07-2015', '20150708'])
dates
```

```python
# объект period - собственно, период. Ниже преобразование DatetimeIndex в PeriodIndex
dates.to_period('D')
```

```python
# создание TimedeltaIndex, для работы с длительностями (time deltas)
dates - dates[0]
```

```python
# удобное создания набора дат (по умолчанию шаг = 1 день, freq - задание шага)
pd.date_range('2015-07-03', '2015-07-10')
pd.date_range('2015-07-03', periods=8)
pd.date_range('2015-07-03', periods=8, freq='H')

# аналогично для range и timedelta
# pd.period_range()
# pd.timedelta_range()
```

```python
# для Rolling statistics есть .rolling() - что это и зачем - читай документацию
# https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html
```

```python
# пример по работе с временными данными
# данные - число велосипедистов пересёкших мост в Сиэтле
# !curl -o FremontBridge.csv https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD

# parse_dates - "перевод" дат в удобный формат? 
data = pd.read_csv('Data_folder/FremontBridge.csv', index_col='Date', parse_dates=True)

# добавили сумму обоих "входов на мост" и сделали имена столбцов читаемыми
data.columns=['West', 'East']
data['Total'] = data.eval('West + East')
data.dropna().describe()
data.head()
```

```python
# визуализируем данные
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn; seaborn.set()

data.plot()
plt.ylabel('Hourly Bicycle Count');
# слишком "плотный" график, нифига не понятно
```

```python
# чтоб он стал читаемым мы заресемплим временные данные

weekly = data.resample('W').sum()
weekly.plot(style=[':', '--', '-'])
plt.ylabel('Weekly bicicle count')
```

```python
# находим rolling mean (среднее скользящее - функция преобразующая исходный ряд
# в "сглаженный" ряд. Она дает значение для точки = среднему предыдущих точек
# исходного ряда)
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count');

# дальше идет работа с matplotlib, и ну её нахрен
# для этого урока хватит
```
