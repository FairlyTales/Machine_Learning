```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
plt.rcParams["figure.figsize"] = (20,13)
```

До этого мы рассматривали supervised learning, а сейчас начнём знакомиться с unsupervised learning.

И начнём с одного из самых популярных алгоритмов этого класса - Principal Component Analysis (PCA), который, по сути, является dimensionality reduction algorithm. Однако может быть использован и для визуализации, фильтрации шумов, выделения features и других вещей.

```python
# посмотрим на "суть" этого алгоритма - уменьшение размерности данных
# для этого создадим данные
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T # np.dot создаёт точки на основе двух массивов
plt.scatter(X[:, 0], X[:, 1]);
```

# Что делают unsupervised learning algorithms

Как можно видеть, данные распределенны практически линейно. И линейная регрессия может построить по ним довольно точную модель предсказания. Однако у РСА задача другая, PCA не предсказывает $y$ по $x$, РСА ищет связи между $y$ и $x$.

In principal component analysis, this relationship is quantified by finding a list of the $principal$ $axes$ in the data, and using those axes to describe the dataset. Using Scikit-Learn's PCA estimator, we can compute this as follows:

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
print(pca.fit(X), '\n')
# PCA.fit() learns some quantities from the data, most importantly the 
# "components" and "explained variance" которые мы сейчас выведем:

print(pca.components_, '\n')
print(pca.explained_variance_)


# однако эти сухие числа нам ничего не говорят, поэтому мы их визуализируем, приняв
# "components" за направление векторов и "explained variance" за длину векторов

# функция для отрисовки найденных 𝑝𝑟𝑖𝑛𝑐𝑖𝑝𝑎𝑙 𝑎𝑥𝑒𝑠 
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# визуализируем исходные данные и полученные principal axes 
plt.scatter(X[:, 0], X[:, 1], alpha=0.4)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

These vectors represent the principal axes of the data, and the length of the vector is an indication of how "important" that axis is in describing the distribution of the data—more precisely, it is a measure of the variance of the data when projected onto that axis. The projection of each data point onto the principal axes are the "principal components" of the data.

# PCA for dimensionaly reduction

Using PCA for dimensionality reduction involves zeroing out one or more of the smallest principal components, resulting in a lower-dimensional projection of the data that preserves the maximal data variance. Делается это так:

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)

# визуализируем результат "уменьшения размерности" данных
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.4)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal');
```

This makes clear what a PCA dimensionality reduction means: the information along the least important principal axis or axes is removed, leaving only the component(s) of the data with the highest variance. The fraction of variance that is cut out (proportional to the spread of points about the line formed in this figure) is roughly a measure of how much "information" is discarded in this reduction of dimensionality.

This reduced-dimension dataset is in some senses "good enough" to encode the most important relationships between the points: despite reducing the dimension of the data by 50%, the overall relationship between the data points are mostly preserved.

# PCA for visualization (на примере Hand-written digits)

Уменьшение размерности с 2х до 1й плоскости (как в примере выше) особого смысла не имеет (мы сделали это только для демонстрации РСА). Однако с данными больших размерностей это очень полезно. Применим РСА для Hand-written digits.

```python
# загрузим данные
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape) # размерность исходных данных равна 64, визуализировать это нереально

# применим РСА
pca = PCA(n_components=2)  # n_components = во сколько "плоскостей" трансформировать данные
projected = pca.fit_transform(digits.data)
print(projected.shape)

# теперь мы можем визуализировать эти данные на двумерной плоскости
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, alpha=0.9,
            cmap='Spectral')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

Recall what these components mean: the full data is a 64-dimensional point cloud, and these points are the projection of each data point along the directions with the largest variance. Essentially, we have found the optimal stretch and rotation in 64-dimensional space that allows us to see the layout of the digits in two dimensions, and have done this in an unsupervised manner—that is, without reference to the labels.

## What do the components of PCA mean?

We can go a bit further here, and begin to ask what the reduced dimensions mean. This meaning can be understood in terms of combinations of basis vectors. For example, each image in the training set is defined by a collection of 64 pixel values, which we will call the vector $x$:

$$ x = [x_1, x_2, x_3 \cdots x_{64}] $$

One way we can think about this is in terms of a pixel basis. That is, to construct the image, we multiply each element of the vector by the pixel it describes, and then add the results together to build the image:

$$ {\rm image}(x) = x_1 \cdot{\rm (pixel~1)} + x_2 \cdot{\rm (pixel~2)} + x_3 \cdot{\rm (pixel~3)} \cdots x_{64} \cdot{\rm (pixel~64)} $$

One way we might imagine reducing the dimension of this data is to zero out all but a few of these basis vectors. For example, if we use only the first eight pixels, we get an eight-dimensional projection of the data, but it is not very reflective of the whole image: we've thrown out nearly 90% of the pixels! And our result will be terrible.



But the pixel-wise representation is not the only choice of basis. We can also use other basis functions, which each contain some pre-defined contribution from each pixel, and write something like

$$ image(x) = {\rm mean} + x_1 \cdot{\rm (basis~1)} + x_2 \cdot{\rm (basis~2)} + x_3 \cdot{\rm (basis~3)} \cdots $$

PCA can be thought of as a process of choosing optimal basis functions, such that adding together just the first few of them is enough to suitably reconstruct the bulk of the elements in the dataset. The principal components, which act as the low-dimensional representation of our data, are simply the coefficients that multiply each of the elements in this series.

Unlike the pixel basis, the PCA basis allows us to recover the salient features of the input image with just a mean plus eight components! The amount of each pixel in each component is the corollary of the orientation of the vector in our two-dimensional example. This is the sense in which PCA provides a low-dimensional representation of the data: it discovers a set of basis functions that are more efficient than the native pixel-basis of the input data.

# Choosing the number of components

Главный вопрос при использовании РСА - сколько нам нужно компонентов для достаточно точного описания данных? Это можно определить посмотрев на cumulative explained variance ratio as a function of the number of components:

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

This curve quantifies how much of the total, 64-dimensional variance is contained within the first $N$ components. For example, we see that with the digits the first 10 components contain approximately 75% of the variance, while you need around 50 components to describe close to 100% of the variance.

Here we see that our two-dimensional projection loses a lot of information (as measured by the explained variance) and that we'd need about 20 components to retain 90% of the variance. Looking at this plot for a high-dimensional dataset can help you understand the level of redundancy present in multiple observations.

# PCA for Noise Filtering

Идея РСА в фильтрации шумов заключается в том что любой комонент с дисперсией значительно превышающей дисперсию шума должен быть unaffected by the noise. Поэтому если мы восстановим данные используя только самые большие subsets of principal components, мы сможем получить нужный нам сигнал и отбросить шум.

Рассмотрим это на примере наших любимых чисел:

```python
# для начала отобразим исходные данные которые не имеют шума
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))
plot_digits(digits.data)

# теперь добавим шум и отобразим полученные данные
np.random.seed(42)
noisy = np.random.normal(digits.data, 4)
plot_digits(noisy)
```

```python
# обучим РСА на шумных данных, задав "сохранение" 50% исходной дисперсии
pca = PCA(0.5).fit(noisy)
print(pca.n_components_) # сколько получилось principal components при сохранении 50% исходной дисперсии
# это чем-то похоже на размытие в фотошопе (а скорее всего это оно и есть)

# визуализируем "очищенные" данные 
components = pca.transform(noisy)
filtered = pca.inverse_transform(components) # .inverse_transform - Transform data back to its original space.
plot_digits(filtered)
```

This signal preserving/noise filtering property makes PCA a very useful feature selection routine—for example, rather than training a classifier on very high-dimensional data, you might instead train the classifier on the lower-dimensional representation, which will automatically serve to filter out random noise in the inputs.

# Пример - Eigenfaces

Earlier we explored an example of using a PCA projection as a feature selector for facial recognition with a support vector machine (пара глав назад - Support Vector Machines). Here we will take a look back and explore a bit more of what went into that. Recall that we were using the Labeled Faces in the Wild dataset made available through Scikit-Learn:

```python
# загрузим сет
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

```python
# зафитим данные на РСА модель
from sklearn.decomposition import PCA
pca = PCA(n_components=150) # размерность исходных данных около 3000, переводим в 150
pca.fit(faces.data)

# визуализируем изображения ассоциирующиеся с несколькими первыми principal components
# (these components are technically known as "eigenvectors," so these types of images
# are often called "eigenfaces"). И они выглядят крипово.
fig, axes = plt.subplots(3, 8, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
```

Результаты позволяют сделать вывод о том как изображения отличаются: the first few eigenfaces (from the top left) seem to be associated with the angle of lighting on the face, and later principal vectors seem to be picking out certain features, such as eyes, noses, and lips. Let's take a look at the cumulative variance of these components to see how much of the data information the projection is preserving:

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

Как видно, "сжатие" размерности до 150 сохраняет более 90% исходной дисперсии. Т.е. используя эти 150 компонентом мы можем востановить ключевые характеристики исходных данных.

```python
# проверим это востановив данные из "сжатого" вида и сравнив их с исходными

# Compute the components and projected faces
pca = PCA(150).fit(faces.data)
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)

# Plot the results
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('150-dim\nreconstruction');
```

Как можно видеть, данные восстановленные из "сжатых" до размерности 150 очень близки к исходным, и мы можем распознать эти лица своими глазами. При этом мы уменьшили их размерность почти в 20 раз!

Это означает, что алгоритм который мы будем использовать для предсказания/классификации этих лиц может обучаться на данных с размерностью 150, а не 3000, без заметных потерь в точности. И такое уменьшение размерности значительно повысит скорость работы алгоритма.

РСА можно использовать для решения большого количества задач: dimensionality reduction, for visualization of high-dimensional data, for noise filtering, and for feature selection within high-dimensional data.
