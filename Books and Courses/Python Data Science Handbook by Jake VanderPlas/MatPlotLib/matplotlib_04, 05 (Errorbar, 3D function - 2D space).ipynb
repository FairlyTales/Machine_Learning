# Глава 4. Errorbar

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.rand(50)

plt.errorbar(x, y, yerr=dy, fmt='.k', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);

# fmt - агрумент отвечающий за внешний вид линий и маркеров
# yerr - errorbar по оси х, еще есть xerr - errorbar по оси х
# ecolor - цвет "линий ошибки", elinewidth - толщина этих линий
# остальное читай в документации
```

# Глава 5. 3D функция в 2D просранстве, plt.contour, contourf, imshow()

```python
# функция которую будем визуализировать
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

# визуалировать будем через plt.contour() - метод рисующий contour plot
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

# np.meshgrid -этот метод создает двухмерные "сетки" из двух одномерных массивов
# это нужно для "присваивания" значений нашей функции, чтоб затем её визуализировать
# по сути, он берет значения из введеных массивов и присваивает их точкам как
# координаты, создавая сетку
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.style.use('dark_background')
plt.figure(figsize=(20, 10))
plt.contour(X, Y, Z, levels=20, cmap='RdGy')
plt.colorbar();
# levels - число уровней/линий равноудаленных друг от друга (по нашим данным)
# RgGy - short for Red-Gray

# полный список colormap есть в plt.cm.<TAB>
```

```python
# если хотим график не из линий, а с "заливкой" - есть plt.contourf()

plt.figure(figsize=(20, 10))
plt.contourf(X, Y, Z, levels=20, cmap='RdGy')
plt.colorbar();
```

```python
# если хотим чтобы график был с плавными цветовыми переходами (без "ступенек")
# есть plt.imshow(). Еще можно увеличить levels в plt.contourf, однако это
# неэффективно с точки зрения объема необходимых вычислений

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
# границы осей это extent=[xmin, xmax, ymin, ymax]
# origin - начало координат, по умолчанию - слева вверху, мы назначаем на
# лево низ, как того требует gridded data (чтоб график был верным)

plt.colorbar()
plt.axis(aspect='image');
# aspect='image' - делаем деления осей Х и У равными
```

```python
# комбинация image plot и contour plot, чтоб нанести на рисунок линии с 
# точными и подписанными значениями

contours = plt.contour(X, Y, Z, levels=3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
# plt.clabel создает подписи на графиках, inline "вписывает" число в линию,
# прерывая её.

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)
# alpha - прозрачность
plt.colorbar();

# использование этих трех методов позволяет отобразить почти любую 3д информацию
# на 2д графике. Дополнительная информация в документации
```
