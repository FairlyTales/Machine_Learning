# Простые графики

```python
# "предстартовая подготовка" матплотлиба

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure() # создание фигуры - нашего "рисунка" в который войдет все остальное
ax = plt.axes() # создание осей

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x)); # использование ООП интерфейса
plt.plot(x, np.sin(x)); # использование матлабовского интерфейса

# для одной фигуры с несколькими графиками вызываем .plot несколько раз
plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
# ; в конце убирает надпись [<matplotlib.lines.Line2D at 0x29a80ff25f8>]
```

# Глава 2. Цвета и стили

```python
# способы задания цветов

plt.plot(x, np.sin(x - 0), color='blue')       # specify color by name
plt.plot(x, np.sin(x - 1), color='g')          # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')       # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')    # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3))# RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse');# all HTML color names supported
```

```python
# варианты типов линий

# полное задание стиля
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# краткое задание стиля
plt.plot(x, x + 4, linestyle='-')   # solid
plt.plot(x, x + 5, linestyle='--')  # dashed
plt.plot(x, x + 6, linestyle='-.')  # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
# цвета и стили можно задавать одновременно, однако не надо так делать
# пиши полностью linestyle=..., color=...; это будет удобнее читать

plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red

# есть и другие способы изменения внешнего вида графиков - читай документацию:)
```

# Задание границ осей
Хотя, они и по умолчанию довольно хороши. Но если нужно запилить что-то нестандартное:

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);

# можно "развернуть" оси задав границы в обратном порядке
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);

# можно задать границы одной строкой при помощи метода .axis (не путать с .axes)
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);

# это же метод позволяет "сжать" границы, чтобы график "вписывался" в них
plt.plot(x, np.sin(x))
plt.axis('tight');

# и сделать так чтобы одна единица по Х была равна одной единице по У
plt.plot(x, np.sin(x))
plt.axis('equal');

# у .axis есть еще несколько удобных функций - читай документацию
```

# Названия осей и графиков, легенда

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve") # задаем имя графика
plt.xlabel("x")           # имя оси Х
plt.ylabel("sin(x)");     # имя оси У
# имена осей можно задать и в .plot через label='...'

# расположение, размер и стиль этих названий можно настроить - читай документацию
```

```python
# создание легенды происходит одной командой - plt.legend()

plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();

# дополнительные настройки легенды читай в документации:)
```

Большинство функций plt напрямую переходят в методы aх (plt.plot() → ax.plot(), plt.legend() → ax.legend(), ...), Однако есть несколько исключений:

plt.xlabel() → ax.set_xlabel()
plt.ylabel() → ax.set_ylabel()
plt.xlim() → ax.set_xlim()
plt.ylim() → ax.set_ylim()
plt.title() → ax.set_title()

```python
# при использовании ООП интерфейса удобно задавать их в методе ax.set()

ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
