```python
# scatter plot задается так же как и обычный график - через .plot()
# однако с третьим аргументом - 'аргумент', который задает тип маркеров

# вот несколько типов маркеров
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend()
plt.xlim(0, 1.8);
```

# plt.scatter

```python
# для создания scatter plots ещё есть специальный метод plt.scatter
# его главное отличие от plt.plot в том, что он может создать график в котором
# свойства каждой отдельной точки могуть быть заданы индивидуально

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
# alpha - настройка прозрачности маркеров

# cmap - задание colormap (A color map (often called a color table or a
# palette) is an array of colors used to map pixel data (represented as
# indexes into the color table) to the actual color values.).

plt.colorbar();  # показать colorbar (который у нас показывает значение цвета)
# в то время как аргумент size представляется в виде размера маркера

# такие графики удобны для представления данных с большим количеством свойств
```

```python
# к примеру представим на plt.scatter датасет Iris, в котором представленны
# три типа цветов у каждого из которых были измерены размеры лепестков и чашелистников

from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

# просто чтобы посмотреть как выглядят данные
import pandas as pd
print(pd.Series(iris), features, iris.feature_names)
# для каждого цветка в data есть 4 значения в списке - длина и ширина
# чашелистника, длина и ширина лепестка. А в target хранятся имена цветков
# 0, 1, 2 = setosa, versicolor, virginica, соответственно

# для каждого маркера задается своё значение
plt.scatter(x=features[0], y=features[1], alpha=0.3, s=100*features[3],
            c=iris.target, cmap='viridis')
# features[0] - первое значение из соответсвующего списка, features[1] - второе
# s=100*features[3] потому что без домножения на 100 маркеры едва видно

plt.colorbar();

# x, y - длинна и ширина чашелистников, размер маркера - толщина лепестков,
# цвет маркера - тип цветка (у нас их три, цвета кажутся разными из-за
# "наслаивания" маркеров с близкими координатами)
```

# Важный момент - хоть plt.scatter и удобен для визуализации данных с большим количество значений, при работе с крупными датасетами (от 1000 маркеров) он начинает тормозить, т.к. каждая точка обрабатывается отдельно.
# В таком случае лучше использовать plt.plot, в котором каждая точка "клонируется". Однако мы теряем возможность задания уникальных параметров для каждой точки.
