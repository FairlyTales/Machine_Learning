# Глава 8. colorbar customization

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])

plt.imshow(I, cmap='viridis') # cmap - определяет цветовое пространство колорбара
# встречается и в других методах plt
plt.colorbar();

# по умолчанию в matplotlib используется cmap='viridis' относящаяся к
# sequential colormaps (которые имеют прямой порядок цветов и хороши для
# отображения множественных данных). Для ситуаций когда нужно отобразить 
# положительные и отрицительные отклонения от среднего/и т.д. стоит использовать
# двойные colorbarы, к примеру RdBu (Red-Blue)
```

```python
# colorbar является одним из элементов plt.Axes, и к нему применимы все способы
# кастомизации plt.Axes. К примеру мы можем установить пределы для colorbar
# что может быть полезно если наше изображение имеет "шум"

# make noise in 1% of the image pixels
speckles = (np.random.random(I.shape) < 0.01)
I[speckles] = np.random.normal(0, 3, np.count_nonzero(speckles))

plt.figure(figsize=(10, 3.5))

plt.subplot(1, 2, 1)
plt.imshow(I, cmap='RdBu')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(I, cmap='RdBu')
plt.colorbar(extend='both') # extend='both' - добавили "треугольники" для "лишних" значений
plt.clim(-1, 1); # задание пределов colorbar

# слева colorbar принимает в расчет "шум", который имеет больший "разлет"
# по границам цвета, и наше изображение становится нечитаемым.
# справа мы вручную задали пределы colorbar, и показали "треугольничками" на
# его концах значения выходящие за его пределы. График стал читаемым.
```

```python
#Дискрентный colorbar

# по умолчанию colorbar непрерывен, но его можно сделать дискретным с помошью
# cmap=plt.cm.get_cmap(), передав этому методу желаемый colorspace и количество
# желаемых значений (корзин/bins). Да, опять костыли

plt.imshow(I, cmap=plt.cm.get_cmap('Blues', 6))
plt.colorbar()
plt.clim(-1, 1);
```

# Глава 9 subplots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
# мы можем создавать оси через plt.axes(), который принимает список из 4х аргументов
# [left, bottom, width, height] которые обозначают себя в СК фигуры

ax1 = plt.axes()  # standard axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2]); # создали дополнительные оси начинающиеся
# на 65% ширины и высоты фигуры, и имеющие "длинну" в 0.2 длины фигуры
```

```python
# через интерфейс ООП матплотбиба создадим стак из двух осей

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4], xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));

# если при попытке создать что-то подобное графики наезжают друг на друга
# то просто пограйся с этими четыремя числами
```

# Subplots

```python
# для создания сразу нескольких графиков рядом друг с другом есть plt.subplot

for i in range(1, 7):
    plt.subplot(2, 3, i) # число строк, столбцов и индекс subplot-а на итоговом рисунке
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
```

```python
# fig.subplots_adjust() позволяет изменить расстояния между сабплотами
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)

# сабплоты через интерфейс ООП - fig.add_subplot()
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
```

```python
# для более удобного создания нескольких сабплотов есть plt.subplots()
# с S на конце, это не то же самое что plt.subplot().
# plt.subplots() создает сетку сабплотов

# rows, columns. sharex='col', sharey='row' - удаляет оси на некоторых сабплотах
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
# plt.subplots() возвращает массив NumPy, с которым удобно работать
# здесь мы обратались к каждому элементу массива и вывели текст на сабплоты

# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')
fig
```

```python
# для более сложных сеток сабплотов есть plt.GridSpec()

# создаем спек для 2х строк, 3х столбцов и желаемой шириной и высотой
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

# затем передаем объект plt.GridSpec методу plt.subplot() и с помошью
# срезов указываем какие входные данные использовать
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

```python
# создание нескольких сабплотов для отображения гистограммы по нескольким осям
# однако с seaborn есть способ сделать это проще, так что не бери в голову

# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
